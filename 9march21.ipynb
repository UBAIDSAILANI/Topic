```python
import pandas as pd
```

```python
import numpy as np
```

```python
Students = {'name': ["Abid", "Ayyaz", "Qasim", "Moiz", "Zeeshan"],
           'Age': [21,25,20,21,23],
           'Python':[79, np.nan,99,78,74],
           'Pandas':[84,78,96, np.nan,83],
           'Numpy':[74,85,96,78,100]}
```

```python
data = pd.DataFrame(Students)
data
```

```python
data['Total_obt']= data['Python']+ data['Pandas']+data['Numpy']
data
```

```python
data.fillna({'Age':data ['Age'].mean(),'Python':0, 'Numpy':0, 'Pandas':0},inplace = True)
data
```

```python
data['Result']=["Pass" if m>=220 else "Try Again" for m in data ['Total_obt']]
```

```python
data
```

```python
data['Result']=["Pass" if m>=150 else "Try Again" for m in data ['Total_obt']]
```

```python
data["Result"] = ["pass" if (python>=50 and numpy>=50 and pandas>=50) else "try again" for python,numpy,pandas in zip (data["Python"],data["Numpy"], data["Pandas"])]
```

```python
data["Percentage"] = [str(np.floor((python+numpy+ pandas)/300*100)) for python,numpy,pandas in zip(data["Python"],data["Numpy"], data["Pandas"])]
```
